```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

ins = datasets.load_iris()
X = ins.data
y= ins.target
print(X[:5])
print(y[:5])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
sc= StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
model = Perceptron(max_iter=40,eta0=0.1,random_state=0)
model.fit(X_train_std,y_train)
y_pred=model.predict(X_test_std)
print(y_pred)
print(y_test)
print('do chinh xac',accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
url = '/crabs.csv'
dataframe = pd.read_csv(url)
dataframe.head(20)
```

```python
dataframe = dataframe.replace('O',1)
dataframe = dataframe.replace('B',0)
dataframe = dataframe.replace('M',1)
dataframe = dataframe.replace("F",0)
dataframe.head(5)
```

```python
X = dataframe.drop(dataframe.columns[[0,1,2]],axis=1)
y= dataframe ['sex']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
sc= StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
print(y_train)
print(X_test_std)
```

```python
model = Perceptron(max_iter=1000,eta0=0.1,random_state=0)
model.fit(X_train_std,y_train)
y_pred=model.predict(X_test_std)
print(y_pred)
print(y_test)
print('do chinh xac',accuracy_score(y_test,y_pred))
```
