```python
from keras.datasets import mnist
import pandas as pd

(train_imgers, train_lables),(test_imgers,test_labels)= mnist.load_data()
X=test_imgers 
Y=test_labels
train_imgers.shape
```

```python
train_lables.shape
test_imgers.shape
test_labels.shape
```

```python
digit = train_imgers[17277]
import matplotlib.pyplot as plt
plt.imshow(digit)
plt.show
print ('so nay la ',train_lables[17277])
```

```python
for i in range (15):
  plt.subplot(3,5,i+1)
  plt.tight_layout
  plt.imshow(train_imgers[i+1000],cmap='gray',interpolation = 'None')
  plt.title('so nay la: {}'.format(train_lables[i+1000]))
  plt.xticks([])
  plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(512, activation = 'relu',input_shape=(28*28,)))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
train_imgers = train_imgers.reshape((60000,28*28))
train_imgers = train_imgers.astype('float32')/255
train_imgers.shape
```

```python
test_imgers = test_imgers.reshape((10000,28*28))
test_imgers=test_imgers.astype('float32')/255
test_imgers.shape
```

```python
from keras.utils import to_categorical
train_lables = to_categorical(train_lables,10)
test_labels=to_categorical(test_labels,10)
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_imgers,train_lables,epochs=20,batch_size=128)
test_loss,test_acc=model.evaluate(test_imgers,test_labels)
print('Do mat mat:',test_loss)
print('Do chinh xac:',test_acc)
```

```python
from keras.datasets import cifar10
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
x_train.shape
```

```python
x_test.shape
```

```python
for i in range (15):
  plt.subplot(3,5,i+1)
  plt.tight_layout
  plt.imshow(x_train[i+1000],cmap='Blues',interpolation = 'None')
  plt.title('so nay la: {}'.format(y_train[i+1000]))
  plt.xticks([])
  plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(1024, activation = 'relu',input_shape=(32*32*3,)))
model.add(Dense(1024, activation = 'relu',input_shape=(32*32*3,)))
model.add(Dense(1024, activation = 'relu',input_shape=(32*32*3,)))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
x_train = x_train.reshape((50000,32*32*3))
x_train = x_train.astype('float32')/255
x_train.shape
```

```python
x_test = x_test.reshape((10000,32*32*3))
x_test = x_test.astype('float32')/255
x_test.shape
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train,10)
y_test=to_categorical(y_test,10)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=20,batch_size=128)
test_loss,test_acc=model.evaluate(x_test,y_test)
print('Do mat mat:',test_loss)
print('Do chinh xac:',test_acc)
```
