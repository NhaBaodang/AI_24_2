```python
import pandas as pd
url = '/file/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot=True)
```

```python

data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model

import numpy as np
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print('gia tri intcvcept la:',model.intercept_)
print('gia tri coffient la:',model.coef_)
```

```python
model.score(pd.DataFrame(data['wt']),data['mpg'])
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
y_pred=model.predict(pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'],data['wt']**2]).T
y=data['mpg']
model.fit(X,data['mpg'])
print('he so Intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)
print('do chinh xac mo hinh:',model.score(X,y))
```

```python
range =np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(pd.DataFrame([range,range**2]).T)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = '/file/50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
correlation = data.corr()
print(correlation)
sns.pairplot(data)
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
print('gia tri intcvcept la:',model.intercept_)
print('gia tri coffient la:',model.coef_)
print('do chinh xac',model.score(pd.DataFrame(data['Profit']),data['R&D Spend']))
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
y_pred=model.predict(pd.DataFrame(data['Profit']))
plt.plot(data['Profit'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Profit'],data['Profit']**2]).T
y=data['R&D Spend']
model.fit(X,data['R&D Spend'])
print('he so Intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)
print('do chinh xac mo hinh:',model.score(pd.DataFrame([data['Profit'],data['Profit']**2]).T,data['R&D Spend']))
range =np.arange(1.5,200000,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='blue',xlim=(25000,200000))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Profit'],data['Profit']**2,data['Profit']**3,data['Profit']**4,data['Profit']**5,data['Profit']**6,data['Profit']**7,data['Profit']**8,data['Profit']**9]).T
y=data['R&D Spend']
model.fit(X,data['R&D Spend'])
print('he so Intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)
print('do chinh xac mo hinh:',model.score(X,data['R&D Spend']))
range =np.arange(1.5,200000,0.1)
X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='blue',xlim=(25000,200000))
plt.plot(range,y_pred,color='blue')
```

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = '/file/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = '/file/Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
sns.heatmap(correlation,annot=True)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('gia tri intcvcept la:',model.intercept_)
print('gia tri coffient la:',model.coef_)
print('do chinh xac',model.score(pd.DataFrame(data['TV']),data['sales']))
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
y_pred=model.predict(pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['TV','Unnamed: 0','radio','newspaper']], y=data['sales'])
print('he thong coeffient',model.coef_)
print('he so intercept',model.intercept_)
print('do chinh xac',model.score(data.loc[:,['TV','Unnamed: 0','radio','newspaper']],data['sales']))
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4]).T
y=data['sales']
model.fit(X,data['sales'])
print('he so Intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)
print('do chinh xac mo hinh:',model.score(X,data['sales']))
range =np.arange(1.5,300,0.1)
X_range = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='blue',xlim=(0,300))
plt.plot(range,y_pred,color='blue')
```
