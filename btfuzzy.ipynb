```python
from skfuzzy.control.controlsystem import ControlSystemSimulation
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
ws= ctrl.Antecedent(np.arange(0.7,3.1,0.1),'WS')
tp = ctrl.Antecedent(np.arange(20.4,41.2,0.1),'TP')
rf = ctrl.Consequent(np.arange(0.2,9.1,0.1),'RF')

ws['verylow_ws']=fuzz.trimf(ws.universe,[0.7,0.7,1.0])
ws['low_ws']=fuzz.trimf(ws.universe,[0.7,1.0,2.0])
ws['normal_ws']=fuzz.trimf(ws.universe,[1.0,2.0,2.7])
ws['high_ws']=fuzz.trimf(ws.universe,[2.0,2.7,3.1])
ws['veryhigh_ws']=fuzz.trimf(ws.universe,[2.7,3.1,3.1])

tp['verylow_tp']=fuzz.trimf(tp.universe,[20.4,20.4,25.6])
tp['low_tp']=fuzz.trimf(tp.universe,[20.4,25.6,29])
tp['normal_tp']=fuzz.trimf(tp.universe,[25.6,29,34.4])
tp['high_tp']=fuzz.trimf(tp.universe,[29,34.4,41.2])
tp['veryhigh_tp']=fuzz.trimf(tp.universe,[34.4,41.2,41.2])

rf['verylow_rf']=fuzz.trimf(rf.universe,[0.2,0.2,2.0])
rf['low_rf']=fuzz.trimf(rf.universe,[0.2,2.0,4.4])
rf['normal_rf']=fuzz.trimf(rf.universe,[2.0,4.4,7.6])
rf['high_rf']=fuzz.trimf(rf.universe,[4.4,7.6,9.1])
rf['veryhigh_rf']=fuzz.trimf(rf.universe,[7.6,9.1,9.1])


luat1 = ctrl.Rule(ws['verylow_ws'] & tp['verylow_tp'],rf['verylow_rf'])
luat2 = ctrl.Rule(ws['verylow_ws'] & tp['low_tp'],rf['verylow_rf'])
luat3 = ctrl.Rule(ws['verylow_ws'] & tp['normal_tp'],rf['low_rf'])
luat4 = ctrl.Rule(ws['verylow_ws'] & tp['high_tp'],rf['normal_rf'])
luat5 = ctrl.Rule(ws['verylow_ws'] & tp['veryhigh_tp'],rf['normal_rf'])
luat6 = ctrl.Rule(ws['low_ws'] & tp['verylow_tp'],rf['verylow_rf'])
luat7 = ctrl.Rule(ws['low_ws'] & tp['low_tp'],rf['verylow_rf'])
luat8 = ctrl.Rule(ws['low_ws'] & tp['normal_tp'],rf['low_rf'])
luat9 = ctrl.Rule(ws['low_ws'] & tp['high_tp'],rf['normal_rf'])
luat10 = ctrl.Rule(ws['low_ws'] & tp['veryhigh_tp'],rf['high_rf'])
luat11 = ctrl.Rule(ws['normal_ws'] & tp['verylow_tp'],rf['verylow_rf'])
luat12 = ctrl.Rule(ws['normal_ws'] & tp['low_tp'],rf['low_rf'])
luat13 = ctrl.Rule(ws['normal_ws'] & tp['normal_tp'],rf['normal_rf'])
luat14 = ctrl.Rule(ws['normal_ws'] & tp['high_tp'],rf['high_rf'])
luat15 = ctrl.Rule(ws['normal_ws'] & tp['veryhigh_tp'],rf['high_rf'])
luat16 = ctrl.Rule(ws['high_ws'] & tp['verylow_tp'],rf['normal_rf'])
luat17 = ctrl.Rule(ws['high_ws'] & tp['low_tp'],rf['normal_rf'])
luat18 = ctrl.Rule(ws['high_ws'] & tp['normal_tp'],rf['high_rf'])
luat19 = ctrl.Rule(ws['high_ws'] & tp['high_tp'],rf['high_rf'])
luat20 = ctrl.Rule(ws['high_ws'] & tp['veryhigh_tp'],rf['veryhigh_rf'])
luat21 = ctrl.Rule(ws['veryhigh_ws'] & tp['verylow_tp'],rf['normal_rf'])
luat22 = ctrl.Rule(ws['veryhigh_ws'] & tp['low_tp'],rf['high_rf'])
luat23 = ctrl.Rule(ws['veryhigh_ws'] & tp['normal_tp'],rf['high_rf'])
luat24 = ctrl.Rule(ws['veryhigh_ws'] & tp['high_tp'],rf['veryhigh_rf'])
luat25 = ctrl.Rule(ws['veryhigh_ws'] & tp['veryhigh_tp'],rf['veryhigh_rf'])



tipping_ctrl = ctrl.ControlSystem([luat1,luat2,luat3,luat4,luat5,luat6,luat7,luat8,luat9,luat10,luat11,luat12,luat13,luat14,luat15,luat16,luat17,luat18,luat19,luat20,luat21,luat22,luat23,luat24,luat25])
tipping = ControlSystemSimulation(tipping_ctrl)
tipping.input['WS']= 2.0
tipping.input['TP']= 30
tipping.compute()
print(tipping.output['RF'])
rf.view(sim = tipping)
```
